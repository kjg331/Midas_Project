```python
import glob
import pandas as pd
```

## Draft Version(Contents combining in one file)

```python
# Draft Version

pattern = 'C:/Users/midashrm/Documents/Pythonwork/직무기술서/*.xlsx'

# Save all file matches: xlsx_files
xlsx_files = glob.glob(pattern)

# Create an empty list: frames
frames = []

#  Iterate over csv_files
for file in xlsx_files:

    #  Read xlsx into a DataFrame
    df = pd.concat(pd.read_excel(file, sheet_name=None),ignore_index=True)

    # Append df to frames
    frames.append(df)

# Concatenate frames into dataframe
df_sheet = pd.concat(frames)
```

```python
df_sheet_1 = df_sheet[~df_sheet["Unnamed: 2"].isin(["직무 구분", "부서장 (팀장 이상)","중요정보 취급","개인정보취급자","주요 개인정보취급자",
                                          "(개인)정보보호","개발","인프라 운영", "DBA", "데이터관리자","직 무 기 술 서" ])]
df_sheet_1 = df_sheet_1[~df_sheet_1["Unnamed: 1"].isin(["직 무 기 술 서"])]
df_sheet_1 = df_sheet_1[~df_sheet_1["Unnamed: 4"].isin(["관리자(부서장) 성명","최종 업데이트 날짜"])]
df_sheet_1 = df_sheet_1.replace([r'[1-9]',"\ㅇ ","o ","\◎ ","○ ","\- ","\. "],"",regex=True)
```

```python
df_sheet_1 = df_sheet_1[['Unnamed: 6',"Unnamed: 2","Unnamed: 3","Unnamed: 4"]]
df_sheet_1.columns = ["성명","3","4","5"]
df_sheet_1 = df_sheet_1[df_sheet_1["성명"].notnull()|df_sheet_1["3"].notnull()
                        |df_sheet_1["4"].notnull()|df_sheet_1["5"].notnull()]
df_sheet_1 = df_sheet_1.iloc[1:,:]
```

## Final Version(Contents for each worksheets)

```python
# Final Version

pattern = 'C:/Users/midashrm/Documents/Pythonwork/직무기술서/*.xlsx'

# Save all file matches: xlsx_files
xlsx_files = glob.glob(pattern)

frames = []

for file in xlsx_files:
    x1 = pd.ExcelFile(file)
    
    for i in range(1, len(x1.sheet_names)):
        #print("reading sheet{0}".format(i))
        df = x1.parse(x1.sheet_names[i]) 

        df_정보 = (df.loc[3:3, "Unnamed: 2":"Unnamed: 6"]).melt().drop("variable",axis=1).dropna()
        df_직무요약 = (df.loc[5:11,"Unnamed: 2":"Unnamed: 3"]).melt().drop("variable",axis=1).dropna()
        df_세부직무구분 = (df.loc[12:35,"Unnamed: 2":"Unnamed: 2"]).melt().dropna("variable",axis=1).dropna()
        df_세부업무내용 = (df.loc[12:35,"Unnamed: 4":"Unnamed: 4"]).melt().drop("variable",axis=1).dropna()

        df_all = pd.concat([df_정보,df_직무요약,df_세부직무구분,df_세부업무내용],axis=1, ignore_index = True)
        frames.append(df_all)
        i += 1

df_sheet = pd.concat(frames)
```

```python
df_sheet = df_sheet.replace([r'[1-9]',"\ㅇ ","o ","\◎ ","○ ","\- ","\. ",'\ㅇ','\ㅇ'],"",regex=True)
```

```python
직무요약 = df_sheet[[0,1]]
직무요약 = 직무요약[직무요약[0].notnull()|직무요약[1].notnull()]

세무직무구분 = df_sheet[[0,2]]
세무직무구분 = 세무직무구분[세무직무구분[0].notnull()|세무직무구분[2].notnull()]

세부업무내용 = df_sheet[[0,3]]
세부업무내용 = 세부업무내용[세부업무내용[0].notnull()|세부업무내용[3].notnull()]
```

```python
with pd.ExcelWriter("C:/Users/midashrm/Documents/Pythonwork/직무기술서/직무기술서_rowdata.xlsx") as writer:

    직무요약.to_excel(writer,index=False, encoding="euc-kr", sheet_name="직무요약")
    세무직무구분.to_excel(writer,index=False, encoding="euc-kr", sheet_name="세무직무구분")
    세부업무내용.to_excel(writer,index=False, encoding="euc-kr", sheet_name="세부업무내용")
```
