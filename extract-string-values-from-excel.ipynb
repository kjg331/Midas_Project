```python
import pandas as pd
import numpy as np
import re
```

```python
org = pd.read_csv("C:/Users/midashrm/Documents/Pythonwork/인사발령/조직도_0305.csv")
```

```python
org.fillna("-",inplace=True)
```

```python
name = org.iloc[:,0:1]

for i in range(0, len(org.columns)):

    name[i] = org.iloc[:,i].str.extract(r'([가-히]{3,4})')
    i += 1
```

```python
name_list = pd.DataFrame(np.array(name.iloc[:]).reshape(-1,1))
```

```python
name_list = name_list.dropna()
name_list = name_list[name_list[0] != "-"]
name_list.to_csv("C:/Users/midashrm/Documents/Pythonwork/인사발령/조직도_0305_zain.csv", encoding="euc-kr", index=False)
```
