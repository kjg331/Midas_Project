## E&C

```python
import glob    
doc = ''   
data = []

for file in glob.glob('C:/Users/midashrm/Documents/Pythonwork/2013/E&C/*.docx'):        
    
    Text = docx.Document(file) 
    
    Tablelist = []

    for i in range(len(Text.tables)):
        table = Text.tables[i]
    
        keys = None
        row_data = None
        
        for i, row in enumerate(Text.tables[i].rows):
            
            rowtext = (cell.text for cell in row.cells)

            if i == 0:
                keys = tuple(rowtext)
                continue
            
            row_data = dict(zip(keys, rowtext))
            data.append(row_data)
        df = pd.DataFrame(data)
    Tablelist.append(df)
    
df_2013_ENC = df.iloc[3:,6:]
df_2013_ENC.to_excel('C:/Users/midashrm/Documents/Pythonwork/2013/E&C/df_2013_ENC.xlsx')
```
